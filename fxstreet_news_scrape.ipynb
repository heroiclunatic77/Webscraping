```python
import requests
import pandas as pd
from datetime import datetime

# API URL
url = "https://www.fxstreet.com/api/PostApi/GetItems?Take=10&CultureName=en&Page=1&Contenttype=1&CacheRefresh=202502251437"

# Send a GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Parse JSON response
    print(data)  # Display data (you can modify this to extract specific details)
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")


# Base API URL
base_url = "https://www.fxstreet.com/api/PostApi/GetItems"

# Number of pages to scrape
num_pages = 10

# List to store extracted data
news_list = []

# Loop through pages
for page in range(1, num_pages + 1):
    # Construct URL with updated page number
    url = f"{base_url}?Take=10&CultureName=en&Page={page}&Contenttype=1"

    # Send GET request
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()  # Parse JSON response

        # Ensure data is a list
        if isinstance(data, list):
            for news in data:
                version_date_str = news.get("VersionDate")
                try:
                    # Attempt to parse the date string
                    version_date = datetime.strptime(version_date_str, "%Y-%m-%dT%H:%M:%SZ")
                    formatted_date = version_date.strftime("%Y-%m-%d %H:%M:%S") # Example format
                except ValueError:
                    formatted_date = version_date_str # Keep original if parsing fails

                news_list.append({
                    "FeedId": news.get("FeedId"),
                    "VersionDate": formatted_date, # Use the formatted date
                    "Title": news.get("Title"),
                    "Summary": news.get("Summary")
                })
        else:
            print(f"Unexpected format on page {page}: {data}")
    else:
        print(f"Failed to retrieve data from page {page}. Status code: {response.status_code}")

# Convert to DataFrame
df = pd.DataFrame(news_list)

# Display DataFrame
df
```
